```python
import requests
import threading
MAX_THREADS = 1000

def send_request():
    rq = requests.post("https://aov-dev.maiquyen.name.vn/admin", json={"key": "value"})
    print(rq.json())

threads = []
for i in range(MAX_THREADS):
    t = threading.Thread(target=send_request)
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```
